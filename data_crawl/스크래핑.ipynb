## 코스피 상위종목
 - 코스피: 네이버금융
 - 종목명,현재가
 - 거래 상위 항목

```python
res = req.get("https://finance.naver.com/sise/sise_quant.naver")
res
```

```python
soup = bs(res.text, "html.parser")
soup
```

```python
naver_title = soup.select("a.tltle")
naver_pr = soup.select("td.number")
naver_no = soup.select("td.no")
len(naver_no)
```

```python
n=0

while(len(naver_title) > n):
    print(f"{naver_no[n].text}. {naver_title[n].text}, {naver_pr[10*n].text}")
    n = n+1
    
```

```python
from tqdm.notebook import tqdm as tqdm # for문 만 사용가능 (전체 길이가 있어야 가능)
```

```python
title_list = []
price_list = []

for i in naver_title:
    title_list.append(i.text)

for i in tqdm(range(0,1000)):
    if i % 10 == 0:
        price_list.append(naver_pr[i].text)

for i in range(0,10):
    print(f"{i+1}. {title_list[i]}, {price_list[i]}")
```

```python
import pandas as pd ## 판다스 라이브러리
```

```python
len(price_list), len(title_list) ## 길이가 같아야함
```

```python
data = {'종목명' : title_list, '가격정보' : price_list}
```

```python
kospi = pd.DataFrame(data)
kospi
```

```python
kospi.to_excel("코스피지수.xlsx", encoding ='utf-8-sig')
```

## 벅스차트 Top 100
 - 순위, 곡, 아티스트 수집

```python
res = req.get("https://music.bugs.co.kr/chart")
res
```

```python
soup = bs(res.text, "html.parser")
soup
```

```python
bug_rank = soup.select("div.ranking>strong")
bug_title = soup.select("p.title>a")
bug_artist = soup.select("p.artist>a")

# n=0
# for i in bug_title:
#     bug_title[n] = (i.text).replace("\n","")
#     n= n+1
# len(bug_rank), len(bug_title), len(bug_artist)
```

```python
ranking_list =[]
for i in bug_rank:
    ranking_list.append(i.text)
len(ranking_list)
```

```python
title_list = []
for i in bug_title:
    title_list.append(i.text)
len(title_list)
```

```python
artist_check = soup.select("a.artistTitle")

for i in artist_check:
    bug_artist.remove(i)

artist_list = []
for i in bug_artist:
    artist_list.append(i.text)

n=0
while(len(artist_list) > n):
    artist_list[n] = (artist_list[n]).replace("\r\n","")
    n =n+1
    
len(artist_list)
```

```python
data = {"순위" : ranking_list, "곡" : title_list, "아티스트" : artist_list}
bugs = pd.DataFrame(data)
```

```python
bugs.set_index("순위")
```

## URL의 규칙 발견하기
 - 여러페이지에서 원하는 정보를 수집하기 위해서는 해당 페이지들의 URL이 특정한 규칙을 가지고 있는지 살펴보자

```python
res = req.get("https://movie.naver.com/movie/point/af/list.naver?&page=1")
res
```

```python
soup = bs(res.text, "html.parser")
soup
```

```python
movie_title = soup.select("td.title > a.movie.color_b")
movie_score = soup.select("div.list_netizen_score>em")
movie_review = soup.select("td.title")
```

```python
title_list = []
for i in movie_title:
    title_list.append(i.text)
len(title_list)
```

```python
score_list =[]
for i in movie_score:
    score_list.append(i.text)
len(score_list)
```

```python
view_list = []
for i in movie_review:
    review = i.text
    review = review.split("\n\n")[2]
    review = review.replace("\t","")
    review = review.replace("\n","")
    review = review.replace("신고","")
    view_list.append(review)
len(view_list)
```

```python
# from tqdm.notebook import tqdm as tqdm 
page = range(1,11)

title_list = []
score_list =[]
view_list = []

for n in tqdm(page):
    res = req.get(f"https://movie.naver.com/movie/point/af/list.naver?&page={n}")
    soup = bs(res.text, "html.parser")
    
    #제목
    movie_title = soup.select("td.title > a.movie.color_b")
    for i in movie_title:
        title_list.append(i.text)
    
    #평점
    movie_score = soup.select("div.list_netizen_score>em")
    for i in movie_score:
        score_list.append(i.text)
    
    #리뷰
    movie_review = soup.select("td.title")
    for i in movie_review:
        review = i.text
        review = review.split("\n\n")[2]
        review = review.replace("\t","")
        review = review.replace("\n","")
        review = review.replace("신고","")
        view_list.append(review)
```

```python
view_list
```

```python
# import pandas as pd
data = {"영화 제목" : title_list, "평점" : score_list, "리뷰" : view_list}
movie_review = pd.DataFrame(data)
# movie_review.to_excel("movie_review.xlsx", encoding ='utf-8-sig')
```
