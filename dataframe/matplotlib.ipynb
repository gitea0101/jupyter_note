```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=[2,4,6]
plt.plot(y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y, ls='--')
plt.show()
# '-' Solid line
# '--' Dashed line
# '-.' Dash-dot line
# ':' dotted line
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y, ls='--', marker = 'o')
plt.show()
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y, ls='--', marker = 'o', color ='r',
         markerfacecolor ='b')
plt.show()
```

```python
x=np.arange(8)
y=[5,2,3,6,8,7,5,7]

plt.plot(x,y,ls='--',marker='o',color='g')
```

```python
plt.plot(x,y,ls=':',marker='D',color='r',mfc='b')
```

```python
plt.plot(x,y,ls='-.',c='m', marker='<', mec='k', mfc='k', ms=10)
```

```python
plt.plot(x,y,ls='--',lw=5, marker='o', ms=15 , mec='g', mew=5, mfc='k')
plt.xticks([0,3,6]) # x  
plt.yticks([1,5,9]) # y  
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]
plt.plot(x)
plt.plot(y)
plt.plot(z)
plt.show()
```

```python
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend() #  
plt.show()
```
